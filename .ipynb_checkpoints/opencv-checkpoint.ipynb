```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# https://pyimagesearch.com/2015/12/07/local-binary-patterns-with-python-opencv/
# https://www.geeksforgeeks.org/create-local-binary-pattern-of-an-image-using-opencv-python/
# https://docs.opencv.org/4.x/dc/ddd/group__line__descriptor.html
```

```python
def read_img(file_path: str, grayscale: bool = True) -> np.ndarray:
    flags = cv2.IMREAD_GRAYSCALE if grayscale else None
    image = cv2.imread(file_path, flags)

    # Check if image is loaded properly
    if image is None:
        raise ValueError("Image not loaded properly. Check the path and file format.")
    return image
```

```python
# Step 2: Convert the image to a binary format
# Apply binary thresholding
_, binary_image = cv2.threshold(original_image, 0, 255, cv2.THRESH_OTSU)


# Step 3: Display the original and processed images side by side
# Create a figure to hold the subplots
plt.figure(figsize=(30, 10))

# Display the original image
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(original_image, cmap='gray')
plt.axis('off')

# Display the binary image
plt.subplot(1, 3, 2)
plt.title('Binary Image')
plt.imshow(binary_image, cmap='gray')
plt.axis('off')

# Show the images
plt.show()
```

```python
# Display the dialated image

# Define the structuring element (kernel)
ksize = 3
kernel = np.ones((ksize, ksize), np.uint8)

dilated_image = cv2.dilate(binary_image, kernel, iterations=2)


plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 3)
plt.title('Dialated Image')
plt.imshow(dilated_image, cmap='gray')
plt.axis('off')

# Show the images
plt.show()
```

```python
# Высота
n_r_from = 3
n_r_to = 35 + n_r_from # y axis

n_c_from = 3 # x axis
n_c_to = 35 + n_c_from


# Ширина
pattern = dilated_image[
n_r_from:n_r_to,
n_c_from:n_c_to]

# Step 4: Extract nxn block
plt.figure(figsize=(2, 2))

# Display the original image
plt.title('Pattern')
plt.imshow(pattern)
plt.axis('off')
plt.show()
```
